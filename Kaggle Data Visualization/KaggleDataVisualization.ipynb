A scatter plot or hex plot is good for what two types of data? Scatter plots and hex plots work best with a mixture of ordinal categorical and interval data.

What type of data makes sense to show in a stacked bar chart, but not in a bivariate line chart? Nominal categorical data makes sense in a stacked bar chart, but not in a bivariate line chart.

What type of data makes sense to show in a bivariate line chart, but not in a stacked bar chart? Interval data makes sense in a bivariate line chart, but not in a stacked bar chart.

Suppose we create a scatter plot but find that due to the large number of points it's hard to interpret. What are two things we can do to fix this issue? One way to fix this issue would be to sample the points. Another way to fix it would be to use a hex plot.

Um pouco de pr-processamento dos dados

```python
footballers = df.copy()
footballers.head()
```

```python
footballers['Unit'] = df['Value'].str[-1] # obtem a unidade de medida de valor
```

```python
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
```

```python
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
```

```python
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
```

```python
footballers = footballers.assign(Value = footballers['Value (M)'],
                              Position = footballers['Preferred Positions'].str.split().str[0])
```

```python
footballers.head()
```

A maneira mais bvia de plotar muitas variveis  aumentar as visualizaes que temos usado at agora com ainda mais variveis visuais.

Uma varivel visual  qualquer dimenso visual ou marcador que podemos usar para distinguir perceptivamente dois elementos de dados um do outro. Exemplos incluem tamanho, cor, forma e uma, duas e at mesmo trs dimenses.

```python
# Multivariate scatter plots
# #hich type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger
sns.lmplot(x = 'Value', y = 'Overall', hue = 'Position', data = 
           footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg = False)
# fit_reg  para estimar a linha de regresso
# este grfico reporta 3 variveis visuais: o valor (eixo x), o atributo overall (eixo y) e o valor pago (pontos)
```

```python
# o mesmo grfico mas mudando o formato dos pontos
sns.lmplot(x = 'Value', y = 'Overall', markers = ['o', 'x', '*'], hue = 'Position',
           data = footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg=False)
```

```python
# Grouped box plot
# Os atacantes marcam mais pontos em "Agresso" do que os goleiros?
f = (footballers.loc[footballers['Position']
                     .isin(['ST', 'GK'])].loc[:, ['Value', 'Overall', 'Aggression', 'Position']])
f = f[f['Overall'] >= 80]
f = f[f['Overall'] < 85]
f['Aggression'] = f['Aggression'].astype(float)

# In this plot, the horizontal axis encodes the Overall score, the vertical axis encodes the Aggression score,
# and the grouping encodes the Position.
sns.boxplot(x = 'Overall', y = 'Aggression', hue = 'Position', data = f)
```

```python
# Summarization
f = (footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']]
    .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan).dropna()).corr()

sns.heatmap(f, annot = True)
```

```python
# Parallel Coordinates

from pandas.plotting import parallel_coordinates

f = (footballers.iloc[:, 12:17].loc[footballers['Position'].isin(['ST', 'GK'])] 
                .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan) 
                .dropna())

f['Position'] = footballers['Position']
f = f.sample(200)
parallel_coordinates(f, 'Position')
```

### GRAMTICA DOS GRFICOS

O processo de design de visualizao de dados comea com a criao de uma figura (1), ajustando a geometria dessa figura (2) e ajustando a esttica (aesthetic) dessa figura (3).

O elemento Data  uma chamada para ggplot, que preenche os dados no grfico. A Esttica  controlada pela funo aes, que preenche nossas variveis visuais: cores, formas e assim por diante. Finalmente, as layers so funes que adicionam ou modificam a plotagem em si.

```python
from plotnine import *
```

```python
top_wines = reviews[reviews['variety'].isin(reviews['variety'].value_counts().head(5).index)]
```

```python
df = top_wines.head(1000).dropna()
```

```python
# o plot  criado por trs operaes:
# 1. plot inicializado passando os dados para o ggplot
# 2. as variveis de interesse so adicionadas no parmetro aes
# 3. a layer  definida com o tipo geom

ggplot(df) + aes('points', 'price') + geom_point()
```

```python
# Para alterar o grfico acima, basta incluir novos itens, como linha de regresso:
(ggplot(df) 
    + aes('points', 'price') 
    + geom_point()
    + stat_smooth())
```

modificar a sada do plot  to simples quanto adicionar um mtodo  cadeia. Como cada modificao que  independente, pode-se fazer essas alteraes em qualquer lugar

```python
# adicionar cores
(ggplot(df) 
    + aes('points', 'price') 
    + aes(color = 'points')
    + geom_point()
    + stat_smooth())
```

```python
# grficos lado a lado
(ggplot(df) 
    + aes('points', 'price') 
    + aes(color = 'points')
    + geom_point()
    + stat_smooth()
    + facet_wrap('variety'))
```

```python
# aes can also appear as a layer parameter:
ggplot(df) + geom_point(aes('points', 'price'))
```

```python
# Or as a parameter in the overall data
ggplot(df, aes('points', 'price')) + geom_point()
```

```python
# bar plot
ggplot(top_wines) + aes('points') + geom_bar()
```

```python
# The plotnine equivalent of a hexplot, a two-dimensional histogram, is geom_bin2d
ggplot(top_wines) + aes('points', 'variety') + geom_bin2d(bins = 20)
```

```python
# adding coord_fixed and ggtitle

(ggplot(top_wines)
   + aes('points', 'variety')
   + geom_bin2d(bins = 20)
   + coord_fixed(ratio = 1)
   + ggtitle('Top Five Most Coomon Wine Variety'))
```

Atualmente, os dados so sobre dia, mas e se juntssemos todas as datas em anos? Isso  conhecido como reamostragem (resampling). Podemos fazer isso para ajustar o conjunto de dados, gerando um resultado agregado por ano. Para fazer isso em Pandas o mtodo  resample. Existem muitas opes de reamostragem em potencial: usaremos Y, que  a abreviao de "year"

```python
shelter_outcomes['date_of_birth'].value_counts().resample('Y').sum().plot.line()
```

```python
stocks.volume.resample('Y').mean().plot.bar()
```

A lag plot compares data points from each observation in the dataset against data points from a previous observation. So for example, data from December 21st will be compared with data from December 20th, which will in turn be compared with data from December 19th, and so on. For example, here is what we see when we apply a lag plot to the volume (number of trades conducted) in the stock data:

```python
from pandas.plotting import lag_plot

lag_plot(stocks.volume.tail(250))

# Parece que os dias em que o volume  alto esto um pouco correlacionados entre si.
# Um dia de negociaes frenticas sinaliza um pouco que o dia seguinte tambm envolver
# negociaes frenticas.
# Lag plots so teis para verificar se o conjunto de dados possui periodicidade (sazonl), mas, s 
# funcionar com dados de sries temporais strong-case
```

## Autocorrelation plot

Esse grfico vai alm do conceito de periodicidade. O grfico de autocorrelao  uma plotagem de sumarizao multivariada que permite verificar cada periodicidade ao mesmo tempo. Ele faz isso calculando a estatstica de a pontuao de correlao em todos os possveis atrasos no conjunto de dados. O que  conhecido como autocorrelao. Autocorrelao  correlao aplicada ao lag.

Em um grfico de autocorrelao, o atraso (lag) est no eixo x e a pontuao de autocorrelao est no eixo y. Quanto mais a autocorrelao se afasta de 0, maior a influncia do registro seguinte sobre o anterior.

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(stocks.volume)
```

Parece que o volume de atividade de negociao est fracamente correlacionado de forma decrescente com o volume negociado do ano anterior.

No h picos no aleatrios significativos no conjunto de dados, portanto, isso  uma boa evidncia de que no h muito de um padro de srie temporal no volume de atividade comercial ao longo do tempo.
