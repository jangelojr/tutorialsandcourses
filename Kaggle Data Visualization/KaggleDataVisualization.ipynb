A scatter plot or hex plot is good for what two types of data? Scatter plots and hex plots work best with a mixture of ordinal categorical and interval data.

What type of data makes sense to show in a stacked bar chart, but not in a bivariate line chart? Nominal categorical data makes sense in a stacked bar chart, but not in a bivariate line chart.

What type of data makes sense to show in a bivariate line chart, but not in a stacked bar chart? Interval data makes sense in a bivariate line chart, but not in a stacked bar chart.

Suppose we create a scatter plot but find that due to the large number of points it's hard to interpret. What are two things we can do to fix this issue? One way to fix this issue would be to sample the points. Another way to fix it would be to use a hex plot.

In data visualization, summarization works by compressing complex data into simpler, easier-to-plot indicators.

Um pouco de pré-processamento dos dados

```python
footballers = df.copy()
footballers.head()
```

```python
footballers['Unit'] = df['Value'].str[-1] # obtem a unidade de medida de valor
```

```python
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
```

```python
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
```

```python
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
```

```python
footballers = footballers.assign(Value = footballers['Value (M)'],
                              Position = footballers['Preferred Positions'].str.split().str[0])
```

```python
footballers.head()
```

A maneira mais óbvia de plotar muitas variáveis ​​é aumentar as visualizações que temos usado até agora com ainda mais variáveis ​​visuais.

Uma variável visual é qualquer dimensão visual ou marcador que podemos usar para distinguir perceptivamente dois elementos de dados um do outro. Exemplos incluem tamanho, cor, forma e uma, duas e até mesmo três dimensões.

```python
# Multivariate scatter plots
# #hich type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger
sns.lmplot(x = 'Value', y = 'Overall', hue = 'Position', data = 
           footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg = False)
# fit_reg é para estimar a linha de regressão
# este gráfico reporta 3 variáveis visuais: o valor (eixo x), o atributo overall (eixo y) e o valor pago (pontos)
```

```python
# o mesmo gráfico mas mudando o formato dos pontos
sns.lmplot(x = 'Value', y = 'Overall', markers = ['o', 'x', '*'], hue = 'Position',
           data = footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], fit_reg=False)
```

```python
# Grouped box plot
# Os atacantes marcam mais pontos em "Agressão" do que os goleiros?
f = (footballers.loc[footballers['Position']
                     .isin(['ST', 'GK'])].loc[:, ['Value', 'Overall', 'Aggression', 'Position']])
f = f[f['Overall'] >= 80]
f = f[f['Overall'] < 85]
f['Aggression'] = f['Aggression'].astype(float)

# In this plot, the horizontal axis encodes the Overall score, the vertical axis encodes the Aggression score,
# and the grouping encodes the Position.
sns.boxplot(x = 'Overall', y = 'Aggression', hue = 'Position', data = f)
```

```python
# Summarization
f = (footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']]
    .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan).dropna()).corr()

sns.heatmap(f, annot = True)
```

```python
# Parallel Coordinates

from pandas.plotting import parallel_coordinates

f = (footballers.iloc[:, 12:17].loc[footballers['Position'].isin(['ST', 'GK'])] 
                .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan) 
                .dropna())

f['Position'] = footballers['Position']
f = f.sample(200)
parallel_coordinates(f, 'Position')
```

### GRAMÁTICA DOS GRÁFICOS

O processo de design de visualização de dados começa com a criação de uma figura (1), ajustando a geometria dessa figura (2) e ajustando a estética (aesthetic) dessa figura (3).

O elemento Data é uma chamada para ggplot, que preenche os dados no gráfico. A Estética é controlada pela função aes, que preenche nossas variáveis visuais: cores, formas e assim por diante. Finalmente, as layers são funções que adicionam ou modificam a plotagem em si.

```python
from plotnine import *
```

```python
top_wines = reviews[reviews['variety'].isin(reviews['variety'].value_counts().head(5).index)]
```

```python
df = top_wines.head(1000).dropna()
```

```python
# o plot é criado por três operações:
# 1. plot inicializado passando os dados para o ggplot
# 2. as variáveis de interesse são adicionadas no parâmetro aes
# 3. a layer é definida com o tipo geom

ggplot(df) + aes('points', 'price') + geom_point()
```

```python
# Para alterar o gráfico acima, basta incluir novos itens, como linha de regressão:
(ggplot(df) 
    + aes('points', 'price') 
    + geom_point()
    + stat_smooth())
```

modificar a saída do plot é tão simples quanto adicionar um método à cadeia. Como cada modificação que é independente, pode-se fazer essas alterações em qualquer lugar

```python
# adicionar cores
(ggplot(df) 
    + aes('points', 'price') 
    + aes(color = 'points')
    + geom_point()
    + stat_smooth())
```

```python
# gráficos lado a lado
(ggplot(df) 
    + aes('points', 'price') 
    + aes(color = 'points')
    + geom_point()
    + stat_smooth()
    + facet_wrap('variety'))
```

```python
# aes can also appear as a layer parameter:
ggplot(df) + geom_point(aes('points', 'price'))
```

```python
# Or as a parameter in the overall data
ggplot(df, aes('points', 'price')) + geom_point()
```

```python
# bar plot
ggplot(top_wines) + aes('points') + geom_bar()
```

```python
# The plotnine equivalent of a hexplot, a two-dimensional histogram, is geom_bin2d
ggplot(top_wines) + aes('points', 'variety') + geom_bin2d(bins = 20)
```

```python
# adding coord_fixed and ggtitle

(ggplot(top_wines)
   + aes('points', 'variety')
   + geom_bin2d(bins = 20)
   + coord_fixed(ratio = 1)
   + ggtitle('Top Five Most Coomon Wine Variety'))
```

## Time-series plotting

```python
# In the "strong case" dates act as an explicit index on your dataset
stocks = pd.read_csv('nyse/prices.csv', parse_dates=['date'])

stocks = stocks[stocks['symbol'] == "GOOG"].set_index('date')

stocks.head()
```

```python
# In the "weak case", dates act as timestamps: they tell us something about when an observation occurred
shelter_outcomes = pd.read_csv(
    'austin-animal-center-shelter-outcomes-and/aac_shelter_outcomes.csv', 
    parse_dates=['date_of_birth', 'datetime'])

shelter_outcomes = shelter_outcomes[
    ['outcome_type', 'age_upon_outcome', 'datetime', 'animal_type', 'breed', 
     'color', 'sex_upon_outcome', 'date_of_birth']]

shelter_outcomes.head()
```

Os dados de stock são agregados em um determinado período de tempo, portanto, alterar o horário altera significativamente os dados. No caso de resultados de animais, a informação é "record-level", as datas são fatos descritivos e não faz sentido alterá-las.

```python
# here's a line plot visualizing which birth dates are the most common in the dataset
shelter_outcomes['date_of_birth'].value_counts().sort_values().plot.line()
```

Atualmente, os dados são sobre dia, mas e se juntássemos todas as datas em anos? Isso é conhecido como reamostragem (resampling). Podemos fazer isso para ajustar o conjunto de dados, gerando um resultado agregado por ano. Para fazer isso em Pandas o método é resample. Existem muitas opções de reamostragem em potencial: usaremos Y, que é a abreviação de "year"

```python
shelter_outcomes['date_of_birth'].value_counts().resample('Y').sum().plot.line()
```

```python
stocks.volume.resample('Y').mean().plot.bar()
```

A lag plot compares data points from each observation in the dataset against data points from a previous observation. So for example, data from December 21st will be compared with data from December 20th, which will in turn be compared with data from December 19th, and so on. For example, here is what we see when we apply a lag plot to the volume (number of trades conducted) in the stock data:

```python
from pandas.plotting import lag_plot

lag_plot(stocks.volume.tail(250))

# Parece que os dias em que o volume é alto estão um pouco correlacionados entre si.
# Um dia de negociações frenéticas sinaliza um pouco que o dia seguinte também envolverá
# negociações frenéticas.
# Lag plots são úteis para verificar se o conjunto de dados possui periodicidade (sazonl), mas, só 
# funcionará com dados de séries temporais strong-case
```

## Autocorrelation plot

Esse gráfico vai além do conceito de periodicidade. O gráfico de autocorrelação é uma plotagem de sumarização multivariada que permite verificar cada periodicidade ao mesmo tempo. Ele faz isso calculando a estatística de a pontuação de correlação em todos os possíveis atrasos no conjunto de dados. O que é conhecido como autocorrelação. Autocorrelação é correlação aplicada ao lag.

Em um gráfico de autocorrelação, o atraso (lag) está no eixo x e a pontuação de autocorrelação está no eixo y. Quanto mais a autocorrelação se afasta de 0, maior a influência do registro seguinte sobre o anterior.

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(stocks.volume)
```

Parece que o volume de atividade de negociação está fracamente correlacionado de forma decrescente com o volume negociado do ano anterior.

Não há picos não aleatórios significativos no conjunto de dados, portanto, isso é uma boa evidência de que não há muito de um padrão de série temporal no volume de atividade comercial ao longo do tempo.
